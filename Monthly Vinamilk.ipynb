```python
def process_data(df, columns, nganh_hang, year):
    filtered_df = df[(df["Ngành hàng"] == nganh_hang) & (df["Thời gian"].str.contains(year, na=False))]
    
    grouped_df = filtered_df.groupby(["Sản phẩm"]).agg(columns).sort_values(by="Total new", ascending=False).reset_index().round()
    
    top_13 = grouped_df.head(13)
    khac = grouped_df.iloc[13:].agg(columns).to_frame().transpose()
    
    result_df = pd.concat([top_13, khac], ignore_index=True)
    result_df.loc[13, 'Sản phẩm'] = 'Sản phẩm khác'
    
    return result_df


# Define columns for Sentiment
sentiment_columns = { "Total new": "sum", "Positive new": "sum", "Neutral new": "sum", "Negative new": "sum"}

# Define nganh_hang values
nganh_hang_list = ['SC', 'SN', 'SB','SĐ','SĐN']

existing_excel_path = r"C:\Users\Kompa-KTelt056\Desktop\Data Kompa\Vinamilk\final.xlsx"
existing_excel = pd.ExcelFile(existing_excel_path)

# Create Excel writer
with pd.ExcelWriter(existing_excel_path, engine='openpyxl', mode='a') as writer:
    for nganh_hang in nganh_hang_list:
        sentiment_data = process_data(sh_sentiment, sentiment_columns, nganh_hang, year)
        
        # Display the sentiment_data
        print(f"Sentiment ({nganh_hang}):")
        print(sentiment_data)
        print("\n" + "="*50 + "\n")  # Add a separator for better visibility
        
        # Save sentiment_data to the existing Excel file
        sheet_name = f"Sentiment_{nganh_hang}"
        sentiment_data.to_excel(writer, sheet_name=sheet_name, index=False)

# Close the existing Excel file
existing_excel.close()


# Define columns for Relevant
relevant_columns = {"Relevant new": "sum","Irrelevant new": "sum","Total new": "sum"}

# Define nganh_hang values
nganh_hang_list = ['SC', 'SN', 'SB','SĐ','SĐN']

existing_excel_path = r"C:\Users\Kompa-KTelt056\Desktop\Data Kompa\Vinamilk\final.xlsx"
existing_excel = pd.ExcelFile(existing_excel_path)

# Create Excel writer
with pd.ExcelWriter(existing_excel_path, engine='openpyxl', mode='a') as writer:
    for nganh_hang in nganh_hang_list:
        relevant_data = process_data(sh_relevant, relevant_columns, nganh_hang, year)
        
        # Display the sentiment_data
        print(f"Relevant ({nganh_hang}):")
        print(relevant_data)
        print("\n" + "="*50 + "\n")  # Add a separator for better visibility
        
        # Save sentiment_data to the existing Excel file
        sheet_name = f"Relevant_{nganh_hang}"
        relevant_data.to_excel(writer, sheet_name=sheet_name, index=False)

# Close the existing Excel file
existing_excel.close()
```

```python
def process_data_channel(df, columns, nganh_hang, year):
    filtered_df = df[(df["Ngành hàng"] == nganh_hang) & (df["Thời gian"].str.contains(year, na=False))]
    
    grouped_df = filtered_df.groupby(["Sản phẩm"]).agg(columns).sort_values(by="Total new", ascending=False).reset_index().round()
    
    grouped_df['Khác'] = grouped_df["Total new"] - grouped_df.drop(columns=["Sản phẩm", "Total new"]).sum(axis=1)
    
    top_13 = grouped_df.head(13)
    khac = grouped_df.iloc[13:].agg(columns).to_frame().transpose()
    
    result_df = pd.concat([top_13, khac], ignore_index=True)
    result_df.loc[13, 'Sản phẩm'] = 'Sản phẩm khác'
    result_df.loc[13, 'Khác'] = result_df.loc[13, 'Total new'] - result_df.loc[13, ['Fanpage new', 'FB page new', 'FB user new', 'FB group new', 'Tiktok new', 'Forum new', 'News new', 'Youtube new']].sum()
    result_df = result_df[['Sản phẩm','Fanpage new','FB page new','FB user new','FB group new','Tiktok new','Forum new','News new','Youtube new','Khác','Total new']]
    
    return result_df

# Define columns for Channel
channel_columns = {"Fanpage new": "sum","FB page new": "sum","FB user new": "sum","FB group new": "sum","Tiktok new": "sum","Forum new": "sum","News new": "sum","Youtube new": "sum","Total new": "sum"}

# Define nganh_hang values
nganh_hang_list = ['SC', 'SN', 'SB','SĐ','SĐN']

existing_excel_path = r"C:\Users\Kompa-KTelt056\Desktop\Data Kompa\Vinamilk\final.xlsx"
existing_excel = pd.ExcelFile(existing_excel_path)

# Create Excel writer
with pd.ExcelWriter(existing_excel_path, engine='openpyxl', mode='a') as writer:
    for nganh_hang in nganh_hang_list:
        channel_data = process_data_channel(sh_channel, channel_columns, nganh_hang, year)
        
        # Display the channel_data
        print(f"Channel ({nganh_hang}):")
        print(channel_data)
        print("\n" + "="*50 + "\n")  # Add a separator for better visibility
        
        # Save sentiment_data to the existing Excel file
        sheet_name = f"Channel_{nganh_hang}"
        channel_data.to_excel(writer, sheet_name=sheet_name, index=False)

# Close the existing Excel file
existing_excel.close()
```
